```python
from pathlib import Path
import pandas as pd
```

```python
#import vine_table csv
data = Path('vine_table.csv')
df = pd.read_csv(data)
df
```

```python
#filter Total_votes column to show only those >= 20
votes_df = df[(df["total_votes"] >= 20)]
votes_df
```

```python
#retrieve rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
# create ['result'] column showing helpful_votes / total_votes
votes_df['Result'] = votes_df['helpful_votes']/votes_df['total_votes']

votes_df
```

```python
#retrieve rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
# filter ['result'] column showing rows greater than/equal to .5
votes_filter = votes_df[(votes_df["Result"] >= .5)]
votes_filter
```

```python
#retrieve rows where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
# delete ['result'] column after filtering
del votes_filter['Result']
votes_filter
```

```python
# filter by customers not enrolled in the vine program 
votes_unpaid = votes_filter[(votes_filter['vine'] == "N")]
votes_unpaid
```

```python
# determine the total amount of votes from customers not enrolled in the vine program 
votes_unpaid['review_id'].count()
```

```python
# determine the total amount of votes with 5-star ratings from customers not enrolled in the vine program 
votes_unpaid[(votes_unpaid["star_rating"] == 5.0)].count()
```

```python
# determine the percentage of 5-star votes from customers not enrolled in the vine program 
(votes_unpaid[(votes_unpaid["star_rating"] == 5.0)].count()) / (votes_unpaid.count())
```

```python
# filter by cusomers enrolled in the vine program 
votes_paid = votes_filter[(votes_filter['vine'] == "Y")]
votes_paid
```

```python
# determine the total amount of votes from customers enrolled in the vine program 
votes_paid['review_id'].count()
```

```python
# determine the total amount of votes with 5-star ratings from customers enrolled in the vine program 
votes_paid[(votes_paid["star_rating"] == 5.0)].count()
```

```python
# determine the percentage of 5-star votes from customers enrolled in the vine program 
(votes_paid[(votes_paid["star_rating"] == 5.0)].count()) / (votes_paid.count())
```
